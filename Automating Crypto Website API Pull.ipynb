# Run an API 

```python
# OBJECTIVE: Run an API to set it as dataframe

def api_runner():
    global df

    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest' 
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'df1e38b9-843c-4ad7-8e36-7acde1f6e2a5',
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)

    df2 = pd.json_normalize(data['data'])
    df2['timestamp'] = pd.to_datetime('now')
    df_append = pd.DataFrame(df2)
    df = pd.concat([df,df_append])

    if not os.path.isfile(r"C:\Users\Takeo\OneDrive\Data Analysis Portfolio\Project Portfolio\Crypto API Tutorial Project\CryptoAPIProject.csv"):
        df.to_csv(r"C:\Users\Takeo\OneDrive\Data Analysis Portfolio\Project Portfolio\Crypto API Tutorial Project\CryptoAPIProject.csv",header = 'columns_name')
    else:
        df.to_csv(r"C:\Users\Takeo\OneDrive\Data Analysis Portfolio\Project Portfolio\Crypto API Tutorial Project\CryptoAPIProject.csv",mode = 'a', header = False)
```

```python
# OBJECTIVE: Set a timer for append the data

import os
from time import time
from time import sleep

for i in range(333):
    api_runner()
    print('API Runner completed')
    sleep(600) # Running each 10 min
exit()
```

# Analyze the Dataframe

```python
# OBJECTIVE: Group percent changes 1H, 24HR, 7D, 30D, 60D and 90D

df3 = df.groupby('name',sort = False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d',]].mean()
df3
```

```python
# OBJECTIVE: Stacking precent changes by specific categorize 

df4 = df3.stack()
df4
```

```python
# Checking the data type

type(df4)
```

```python
# OBJECTIVE: Prettify the stacking

df5 = df4.to_frame(name = 'values')
df5
```

```python
# Adding the index into the columns

index = pd.Index(range(90))
df6 = df5.reset_index()
df6
```

```python
df7 = df6.rename(columns = {'level_1':'percent_change'})
df7
```

```python
# Rename the each cells name in the percent change

df7['percent_change'] = df7['percent_change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d'],['1hr','24hr','7d','30d','60d','90d'])
df7
```
